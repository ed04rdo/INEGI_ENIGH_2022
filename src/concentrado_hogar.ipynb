<H5>CONJUNTO DE DATOS</H5>

```python
concentrado_hogar = pd.read_csv(path+"/../input/conjunto_de_datos_concentradohogar_enigh2022_ns/conjunto_de_datos/conjunto_de_datos_concentradohogar_enigh2022_ns.csv",
                                dtype=str)
```

```python
seleccion_concentrado_hogar = concentrado_hogar[["folioviv", "foliohog", "ing_cor", "ingtrab", "trabajo", "negocio","otros_trab", "rentas", 
                                                 "utilidad", "arrenda", "transfer", "jubilacion","becas", "donativos", "remesas", "bene_gob", 
                                                 "transf_hog", "trans_inst","estim_alqu", "otros_ing", "factor", "upm", "est_dis"]]
seleccion_concentrado_hogar.folioviv.nunique()
```

```python
## reordenamiento de acuerdo a ingreso corriente, folio vivienda, folio hogar
seleccion_concentrado_hogar = seleccion_concentrado_hogar.sort_values(by=["ing_cor","folioviv","foliohog"])
## suma de los factores de expansión asignados a cada vivienda
sum_factores = seleccion_concentrado_hogar.factor.astype(int).sum()
## Se divide la suma de factores entre diez para sacar el tamaño del decil (se debe truncar)
decil_factores = sum_factores//10
## Muestra la suma del factor en variable hogar
seleccion_concentrado_hogar["decil_factores"] = decil_factores
```
